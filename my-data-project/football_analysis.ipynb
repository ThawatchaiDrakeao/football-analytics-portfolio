```python
# ⚽️ Football Forward Efficiency Analysis
**[Task 1-4: Setup, Goals & Data Questions]**

**EN:** This project aims to analyze the finishing efficiency of top football forwards by comparing their actual Goals scored against their Expected Goals (xG). The goal is to identify which players overperform their underlying metrics.
**TH:** โปรเจกต์นี้มีเป้าหมายเพื่อวิเคราะห์ความเฉียบคมในการทำประตูของกองหน้าระดับท็อป โดยเปรียบเทียบระหว่าง "ประตูที่ทำได้จริง (Goals)" กับ "โอกาสความน่าจะเป็นในการได้ประตู (xG)" เพื่อค้นหาว่าผู้เล่นคนใดมีประสิทธิภาพการจบสกอร์สูงกว่าค่ามาตรฐาน
```

```python
# [Task 5: Load and Inspect the Data]

import pandas as pd

# EN: Create a dataset for top forwards (Goals vs xG)
# TH: สร้างชุดข้อมูลสถิติของกองหน้าระดับท็อป (ประตูจริง เปรียบเทียบกับ xG)
data = {
    'Player': ['Mohamed Salah', 'Erling Haaland', 'Bukayo Saka', 'Son Heung-min'],
    'Team': ['Liverpool', 'Man City', 'Arsenal', 'Spurs'],
    'Goals': [18, 25, 14, 15],
    'xG': [16.5, 26.2, 11.8, 12.1]
}

# EN: Convert into a Pandas DataFrame
# TH: แปลงข้อมูลให้เป็นตาราง Pandas DataFrame
df = pd.DataFrame(data)

# EN: Inspect the first few rows and data information
# TH: ตรวจสอบข้อมูล 5 แถวแรก และดูรายละเอียดของชุดข้อมูล
print("--- Data Preview / ตัวอย่างข้อมูล ---")
display(df.head())

print("\n--- Data Info / โครงสร้างข้อมูล ---")
df.info()
```

```python
# [Task 6: Data Cleaning]

# EN: Fill missing values with 0 and ensure numeric data types
# TH: เติมค่าว่าง (ถ้ามี) ด้วย 0 และตรวจสอบให้แน่ใจว่าข้อมูลเป็นตัวเลขสำหรับการคำนวณ
df['Goals'] = df['Goals'].fillna(0)
df['xG'] = df['xG'].fillna(0)

df['Goals'] = pd.to_numeric(df['Goals'])
df['xG'] = pd.to_numeric(df['xG'])

# EN: Verify that there are no missing values left
# TH: ยืนยันผลการทำความสะอาดข้อมูลว่าไม่มีค่าว่างหลงเหลืออยู่
print("--- Missing Values Check / ตรวจสอบค่าว่าง ---")
print(df[['Goals', 'xG']].isnull().sum())
```

```python
# [Task 7: Data Preparation]

# EN: Calculate finishing efficiency (Goals - xG). Positive values indicate overperformance.
# TH: สร้างคอลัมน์คำนวณความเฉียบคม (Goals - xG) ค่าบวกหมายถึงการยิงประตูได้ดีกว่าโอกาสที่ควรจะเป็น
df['Finishing_Efficiency'] = df['Goals'] - df['xG']

# EN: Preview the updated DataFrame
# TH: แสดงผลตารางที่เพิ่มคอลัมน์ใหม่แล้ว
display(df[['Player', 'Goals', 'xG', 'Finishing_Efficiency']].head())
```

```python
# [Task 8: Exploratory Data Analysis (EDA)]

# EN: Sort players by Finishing Efficiency in descending order to find the most clinical forward
# TH: จัดเรียงลำดับผู้เล่นตามความเฉียบคมจากมากไปน้อย เพื่อหากองหน้าที่จบสกอร์ได้ดีที่สุด
df_sorted = df.sort_values(by='Finishing_Efficiency', ascending=False)

print("🏆 Top Forwards by Finishing Efficiency / จัดอันดับกองหน้าที่เฉียบคมที่สุด 🏆")
display(df_sorted)
```
